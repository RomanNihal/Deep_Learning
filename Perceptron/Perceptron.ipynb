```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
```

```python
print(df.shape)
df.head()
```

```python
# @title CGPA vs Resume Score

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['cgpa'], df['resume_score'], c=df['placed'], cmap='viridis', alpha=0.7)
plt.xlabel('CGPA')
plt.ylabel('Resume Score')
plt.title('CGPA vs Resume Score')
_ = plt.colorbar(label='Placement Status (0: No, 1: Yes)')
```

```python
sns.scatterplot(x='cgpa', y='resume_score', hue='placed', data=df)
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
p.fit(X,y)
```

```python
print(p.coef_)
print(p.intercept_)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```
